```python
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)

URL = r"https://www.dropbox.com/s/qhhlqcica1nvtaw/opencv_bootcamp_assets_NB1.zip?dl=1"

asset_zip_path = os.path.join('/content/drive/My Drive/Assignment', f"opencv_bootcamp_assets_NB1.zip")

# Download if assest ZIP does not exists.
# if not os.path.exists(asset_zip_path):
download_and_unzip(URL, asset_zip_path)

```

```python
img=cv2.imread('/content/drive/My Drive/Assignment/New_Zealand_Lake.jpg',1)
```

```python
cv2_imshow(img)
```

```python
print(img.shape)
```

```python
# Brightness in the context of image processing refers to the overall intensity or luminance of an image. It represents how much light is emitted or reflected by the objects in the image. Images with higher brightness appear lighter, while those with lower brightness appear darker.

img2=np.ones((img.shape),np.uint8)*250
# creating the image with same as  img but having all pixels values equal to 50
cv2_imshow(img2)
brighter_img=cv2.add(img,img2)
cv2_imshow(brighter_img)
darker=cv2.subtract(img,img2)
# add and subract increase the brightness of image.Add function add the pixel value of 2nd image in 1st imgae and make it brighter while substract function substract the pixels and make the image darker
cv2_imshow(darker)
```

```python
# In the context of image processing, contrast refers to the difference in luminance or color between the darkest and brightest parts of an image.
# Contrast in an image refers to the difference in brightness between objects and the background. By multiplying the intensity values of pixels by a constant factor, you can adjust the contrast. If the multiplying factor is greater than 1, it increases the contrast, making the bright areas brighter and the dark areas darker. Conversely, if the factor is less than 1, it decreases the contrast, bringing the intensity values closer together.

img3=np.ones(img.shape)*1.2
brighter1=np.uint8(np.clip(cv2.multiply(np.float64(img),img3),0,255))
cv2_imshow(brighter1)


img4=np.ones(img.shape)*0.8
darker1=np.uint8(cv2.multiply(np.float64(img),img4))
cv2_imshow(darker1)
```
